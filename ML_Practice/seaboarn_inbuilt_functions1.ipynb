```python
import seaborn as sns
```

```python
df = sns.load_dataset("tips")
df.head()
```

```python
#correlation by default takes pearson methos and min_period as 1
df.corr()
```

```python
df.corr(method='kendall')
```

```python
df.corr(min_periods=5)#Minimum number of observations required per pair of columns to have a valid result
```

```python
sns.heatmap(df.corr())
```

```python
"""
Reference links : 
https://likegeeks.com/seaborn-heatmap-tutorial/
https://seaborn.pydata.org/generated/seaborn.heatmap.html
"""
```

```python
"""
vmax, vmin :
Anchoring the colormap with setting vmax and vmin, color bar on side too changes as per the range.
Also if there are multiple data to be plotted , data between the range will only be displayed.

cmap : Sequential colormap

annot : values of corr dispalyed in ceter

fmt : values of corr displayed till that decimal place( example here upto 3)

xticklables/ yticklabels if false removes that labels

"""
sns.heatmap(df.corr(), vmin=0, vmax=1, cmap='YlGnBu', annot=True, fmt='.3g',yticklabels=False,
            cbar_kws={'label': 'My Colorbar', 'orientation': 'horizontal'})
```

```python
"""
kind : { "scatter" | "reg" | "resid" | "kde" | "hex" }, Kind of plot to draw.
ratio : Ratio of joint axes height to marginal axes height. i.e histpgram vs inside plotted hex

"""
sns.jointplot("total_bill", "tip", data=df, height=5, ratio=12, color="g", kind='hex')
```

```python
sns.pairplot(df, hue ='sex')
```
